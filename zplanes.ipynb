```python
def  rmsv(data):
    RSS = np.sum(np.square(data))
    rse = math.sqrt(RSS / (len(data)))
    return(rse)
```

```python
def ret_class(test_ids,ids_ve,ids_vm_cc,ids_nc,test_lab,test):
    xy, ve_ind, b_ind= np.intersect1d(test_ids, ids_ve, return_indices=True)
    xy, vm_ind, b_ind= np.intersect1d(test_ids, ids_vm_cc, return_indices=True)
    xy, nc_ind, b_ind= np.intersect1d(test_ids, ids_nc, return_indices=True)
    #VE=[e for e in test_ids if e in ids_ve]
    #VM=[e for e in test_ids if e in ids_vm_cc]
    #VC=[e for e in test_ids if e in ids_nc]
    
    ve_vertex_o=test_lab[ve_ind]
    vm_vertex_o=test_lab[vm_ind]
    nc_vertex_o=test_lab[nc_ind]
    #print(len(test_ids))
    #print(len(ids_ve),len(ids_vm_cc),len(ids_nc))
    print(len(ve_ind),len(vm_ind),len(nc_ind))
    ve_vertex_t=test[ve_ind]
    vm_vertex_t=test[vm_ind]
    nc_vertex_t=test[nc_ind]
    #print(len(test_ids),len(ids_ve),len(ids_vm_cc),len(ids_nc))
    re_ve=resid(ve_vertex_t,ve_vertex_o,1)
    re_vm=resid(vm_vertex_t,vm_vertex_o,1)
    re_nc=resid(nc_vertex_t,nc_vertex_o,1)
    re_ve=np.array(re_ve)
    re_vm=np.array(re_vm)
    re_nc=np.array(re_nc)
    return ve_vertex_o,vm_vertex_o,nc_vertex_o,ve_vertex_t,vm_vertex_t,nc_vertex_t,re_ve,re_vm,re_nc
```
